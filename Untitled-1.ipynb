```python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
import numpy as np
from dataset import SubsetCIFAR
classes_to_keep = {0, 1} # set for fast existence check
# Create a new dataset that only keeps the desired classes
```

```python
transform = transforms.Compose([
        transforms.ToTensor(),
        transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
    ])

    # Download and load the training data
trainset = SubsetCIFAR(root='./data', train=True, download=True, indices=classes_to_keep, transform=transform)
trainloader = torch.utils.data.DataLoader(trainset, batch_size=64, shuffle=True, num_workers=2)
print(trainset.targets)
# Download and load the test data
testset = SubsetCIFAR(root='./data', train=False, download=True, indices=classes_to_keep,transform=transform)
testloader = torch.utils.data.DataLoader(testset, batch_size=64, shuffle=False, num_workers=2)
print(testset.targets)
```

```python
class SimpleNN(nn.Module):
    def __init__(self):
        super(SimpleNN, self).__init__()
        self.fc1 = nn.Linear(32*32*3, 128)
        self.fc2 = nn.Linear(128, 1)
        # self.fc3 = nn.Linear(256, 10)
        self.relu = nn.ReLU()
        self.sigmoid = nn.Sigmoid()
    
    def forward(self, x):
        x = x.view(-1, 32*32*3)  # Flatten the image
        x = self.relu(self.fc1(x))
        x = self.sigmoid(self.fc2(x))
        # x = self.fc3(x)
        return x

    # Initialize the model, loss function, and optimizer
model = SimpleNN()
criterion = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.001, momentum=0.9)
```

```python
def train(model, trainloader, criterion, optimizer, epochs=2):
    for epoch in range(epochs):
        running_loss = 0.0
        for inputs,labels in trainloader:
            #inputs, labels = data
            
            optimizer.zero_grad()
            
            outputs = model(inputs)
            # print("op",outputs.squeeze(dim=1))
            # print("label",labels.float())
            loss = criterion(outputs.squeeze(dim=1), labels.float())
            loss.backward()
            optimizer.step()
            
            running_loss += loss.item()
            # if i % 100 == 99:  # print every 100 mini-batches
            #     print(f'Epoch {epoch + 1}, Batch {i + 1}, Loss: {running_loss / 100:.4f}')
            #     running_loss = 0.0

    # Evaluation loop
def evaluate(model, testloader):
    correct = 0
    total = 0
    with torch.no_grad():
        for data in testloader:
            images, labels = data
            outputs = model(images)
            print(outputs)
            print(labels)
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
            break
    
    print(f'Accuracy of the network on the 10000 test images: {100 * correct / total:.2f}%')

```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

n_steps = 10
# Length of singular values
m_singular_values = 20

# Generate a sorted array of length m_singular_values with values between 1 and 100
singular_value_array = np.sort(np.random.uniform(1, 100, m_singular_values))[::-1]

# Repeat this step n times to form the singular_values matrix
singular_values = np.tile(singular_value_array, (n_steps, 1))

# Creating the indices for plotting
steps = np.arange(n_steps)  # Step indices
singular_value_indices = np.arange(m_singular_values)  # Singular value indices

# Creating the 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plotting each step as a line
for i in range(n_steps):
    ax.plot(singular_value_indices, [steps[i]] * m_singular_values, singular_values[i, :], marker='')

ax.set_xlabel('Singular Value Index')
ax.set_ylabel('Step')
ax.set_zlabel('Singular Value')
ax.set_title('Singular Values over Steps')

plt.show()
```
